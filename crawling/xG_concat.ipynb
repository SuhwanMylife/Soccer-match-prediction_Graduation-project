```python
import pandas as pd
import numpy as np
```

```python
# 원 데이터셋 불러오기

loc = "../Datasets/"

raw_data_22 = pd.read_csv(loc + '2014-15.csv')
```

```python
# xG 데이터 불러오기

xG_data_22 = pd.read_csv('xG_2014-15.csv')
```

```python
# 원 데이터셋 column 생성
raw_data_22["Home_xG"] = np.NaN
raw_data_22["Home_xA"] = np.NaN
raw_data_22["Home_xPTS"] = np.NaN
raw_data_22["Away_xG"] = np.NaN
raw_data_22["Away_xA"] = np.NaN
raw_data_22["Away_xPTS"] = np.NaN
```

```python
# 팀 명 수정
for i in range(len(xG_data_22)):
    if xG_data_22.iloc[i][1] == "Manchester United":
        xG_data_22.loc[i, "Team"] = "Man United"
    if xG_data_22.iloc[i][1] == "Wolverhampton Wanderers":
        xG_data_22.loc[i, "Team"] = "Wolves"
    if xG_data_22.iloc[i][1] == "Manchester City":
        xG_data_22.loc[i, "Team"] = "Man City"
    if xG_data_22.iloc[i][1] == "Newcastle United":
        xG_data_22.loc[i, "Team"] = "Newcastle"
    if xG_data_22.iloc[i][1] == "West Bromwich Albion":
        xG_data_22.loc[i, "Team"] = "West Brom"
    if xG_data_22.iloc[i][1] == "Queens Park Rangers":
        xG_data_22.loc[i, "Team"] = "QPR"
```

```python
# xG 데이터 추가 (날짜 형식: 0000-00-00)

for i in range(len(xG_data_22)):
    for j in range(len(raw_data_22)):
        if xG_data_22.iloc[i][0][:6] + xG_data_22.iloc[i][0][8:10] == raw_data_22.iloc[j][1]:
            if xG_data_22.iloc[i][1] == raw_data_22.iloc[j][2]:
                raw_data_22.loc[j, 'Home_xG'] = xG_data_22.iloc[i][2]
                raw_data_22.loc[j, 'Home_xA'] = xG_data_22.iloc[i][3]
                raw_data_22.loc[j, 'Home_xPTS'] = xG_data_22.iloc[i][4]
                continue
            elif xG_data_22.iloc[i][1] == raw_data_22.iloc[j][3]:
                raw_data_22.loc[j, 'Away_xG'] = xG_data_22.iloc[i][2]
                raw_data_22.loc[j, 'Away_xA'] = xG_data_22.iloc[i][3]
                raw_data_22.loc[j, 'Away_xPTS'] = xG_data_22.iloc[i][4]
                continue
raw_data_22
```
